```python
import pandas as pd
import numpy as np
diabetic_data = "C:\\Diabetic data\\diabetic_data.csv"
df = pd.read_csv(diabetic_data)
df
```

```python
print('Melt')
df.melt(id_vars=['patient_nbr', 'encounter_id'], value_vars=['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications'])
```

```python
print('Pivot')
df.pivot(index = ['encounter_id'], columns =  'time_in_hospital', values = 'number_diagnoses')
```

```python
print('Aggregation')
df.agg({'num_procedures' : ['min', 'max'] , 'num_medications' : ['sum', 'min', 'max']})
```

```python
print('Iteration')
for key,value in df.items():
    print(key,value)
```

```python
print('GroupBy')
gender_data = df.groupby('gender')[['time_in_hospital','num_lab_procedures', 'num_procedures', 'num_medications', 'number_diagnoses']].mean()
gender_data
```

```python
age_data = df.groupby('age')[['time_in_hospital','num_lab_procedures', 'num_procedures', 'num_medications', 'number_diagnoses']].mean()
age_data
```

```python
race_data = df.groupby('race')[['time_in_hospital','num_lab_procedures', 'num_procedures', 'num_medications', 'number_diagnoses']].mean()
race_data
```
